# Problem 1. Generate an expression tree (Binary tree) from a postfix expression

def constructBinaryTree(self, input) -> TreeNode:
    operators = {'+', '-', '*', '/'}
    stack = []
    for token in input:
        if token in operators:
            right = stack.pop()
            left = stack.pop()
            node = TreeNode(token, left, right)
            stack.append(node)
        else:
            stack.append(TreeNode(token))
    return stack.pop()

This function constructs a binary expression tree from a postfix (reverse Polish notation) expression provided as a list of string tokens. The resulting tree can then be traversed in different orders to produce prefix, infix, or postfix notations.

## Prefix traversal

def prefixNotationPrint(self, head: TreeNode) -> list:
    result = []
    def preorder(node):
        if node is None:
            return
        result.append(node.val)
        preorder(node.left)
        preorder(node.right)
    preorder(head)
    return result


Generates prefix notation (Polish notation) from the expression tree by performing a pre-order traversal. In prefix notation, the operator appears before its operands.
