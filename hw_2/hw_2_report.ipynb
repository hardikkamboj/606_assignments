# Problem 1. Generate an expression tree (Binary tree) from a postfix expression

def constructBinaryTree(self, input) -> TreeNode:
    operators = {'+', '-', '*', '/'}
    stack = []
    for token in input:
        if token in operators:
            right = stack.pop()
            left = stack.pop()
            node = TreeNode(token, left, right)
            stack.append(node)
        else:
            stack.append(TreeNode(token))
    return stack.pop()

# Problem 3. Evaluate a postfix expression using a stack. That is, implement a function that evaluates a postfix expression without constructing an expression tree.

First, will implement the stack basic functions. 



 class Stack:
 
    def __init__(self):
        self.data = []
        self.top = -1

    def push(self, val):
        self.data.append(val)
        self.top += 1

    def pop(self):
        if self.isEmpty():
            raise IndexError("Stack is empty")
        val = self.data[self.top]
        self.data.pop()
        self.top -= 1
        return val

    def peek(self):
        if self.isEmpty():
            raise IndexError("Stack is empty")
        return self.data[self.top]

    def isEmpty(self):
        return self.top == -1

    def size(self):
        return self.top + 1
